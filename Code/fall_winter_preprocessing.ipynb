```python
import pandas as pd
```

```python
fall = pd.read_csv('C:/Users/LG/Downloads/fall.csv')
winter = pd.read_csv('C:/Users/LG/Downloads/winter.csv')
fall_test = pd.read_csv('C:/Users/LG/Downloads/fall_test.csv')
winter_test = pd.read_csv('C:/Users/LG/Downloads/winter_test.csv')
```

```python
fall
```

```python
fall_test
```

```python
#일몰 이후 0
fall.loc[(fall['hh'] <= fall['sunriseh']) |
         (fall['hh'] >= fall['sunseth']) &
         (fall['si'].isna()), 'si'] = 0

fall_test.loc[(fall_test['hh'] <= fall_test['sunriseh']) |
         (fall_test['hh'] >= fall_test['sunseth']) &
         (fall_test['si'].isna()), 'si'] = 0
```

```python
fall.loc[(fall['hh'] <= fall['sunriseh']) |
         (fall['hh'] >= fall['sunseth']) &
         (fall['ss'].isna()), 'ss'] = 0

fall_test.loc[(fall_test['hh'] <= fall_test['sunriseh']) |
         (fall_test['hh'] >= fall_test['sunseth']) &
         (fall_test['ss'].isna()), 'ss'] = 0
```

```python
# 가을은 결측값 모두 0 
fall.loc[(fall['sn'].isna()), 'sn'] = 0
fall_test.loc[(fall_test['sn'].isna()), 'sn'] = 0
```

```python
#일사, 일조량 평균
simean=fall.loc[fall['ww'] == 'C','si'].mean()
simean2=fall.loc[fall['ww'] != 'C','si'].mean()
ssmean=fall.loc[fall['ww'] == 'C','ss'].mean()
ssmean2=fall.loc[fall['ww'] != 'C','ss'].mean()
```

```python
fall.loc[(fall['ww']=='C') &
         (fall['si'].isna()), 'si'] = simean
fall.loc[(fall['ww']!='C') &
         (fall['si'].isna()), 'si'] = simean2

fall_test.loc[(fall_test['ww']=='C') &
         (fall_test['si'].isna()), 'si'] = simean
fall_test.loc[(fall_test['ww']!='C') &
         (fall_test['si'].isna()), 'si'] = simean2
```

```python
fall.loc[(fall['ww']=='C') &
         (fall['ss'].isna()), 'ss'] = ssmean
fall.loc[(fall['ww']!='C') &
         (fall['ss'].isna()), 'ss'] = ssmean2

fall_test.loc[(fall_test['ww']=='C') &
         (fall_test['ss'].isna()), 'ss'] = ssmean
fall_test.loc[(fall_test['ww']!='C') &
         (fall_test['ss'].isna()), 'ss'] = ssmean2
```

```python
#강수량 / 강수 시간 평균 
rnmean=fall.loc[fall['ww'] == 'R','rn'].mean()
rnmean2=fall.loc[(fall['ww'] == 'F') | (fall['ww'] == 'H'),'rn'].mean()
rnmean3=fall.loc[(fall['ww'] != 'R') & 
                 (fall['ww'] != 'F') & (fall['ww'] != 'H'),'rn'].mean()

remean=fall.loc[fall['ww'] == 'R','re'].mean()
remean2=fall.loc[(fall['ww'] == 'F') | (fall['ww'] == 'H'),'re'].mean()
remean3=fall.loc[(fall['ww'] != 'R') & 
                 (fall['ww'] != 'F') & (fall['ww'] != 'H'),'re'].mean()
```

```python
fall.loc[(fall['ww']=='R') &
         (fall['rn'].isna()), 'rn'] = rnmean
fall.loc[(fall['ww'] == 'F') | (fall['ww'] == 'H') &
         (fall['rn'].isna()), 'rn'] = rnmean2
fall.loc[(fall['ww'] != 'R') & 
         (fall['ww'] != 'F') & (fall['ww'] != 'H') &
         (fall['rn'].isna()), 'rn'] = rnmean3


fall_test.loc[(fall_test['ww']=='R') &
         (fall_test['rn'].isna()), 'rn'] = rnmean
fall_test.loc[(fall_test['ww'] == 'F') | (fall['ww'] == 'H') &
         (fall_test['rn'].isna()), 'rn'] = rnmean2
fall_test.loc[(fall_test['ww'] != 'R') & 
         (fall_test['ww'] != 'F') & (fall['ww'] != 'H') &
         (fall_test['rn'].isna()), 'rn'] = rnmean3
```

```python
fall.loc[(fall['ww']=='R') &
         (fall['re'].isna()), 're'] = remean
fall.loc[(fall['ww'] == 'F') | (fall['ww'] == 'H') &
         (fall['re'].isna()), 're'] = remean2
fall.loc[(fall['ww'] != 'R') & 
         (fall['ww'] != 'F') & (fall['ww'] != 'H') &
         (fall['re'].isna()), 're'] = remean3


fall_test.loc[(fall_test['ww']=='R') &
         (fall_test['re'].isna()), 're'] = remean
fall_test.loc[(fall_test['ww'] == 'F') | (fall['ww'] == 'H') &
         (fall_test['re'].isna()), 're'] = remean2
fall_test.loc[(fall_test['ww'] != 'R') & 
         (fall_test['ww'] != 'F') & (fall['ww'] != 'H') &
         (fall_test['re'].isna()), 're'] = remean3
```

```python
#기온 
tamean=fall['ta'].mean()
fall.loc[fall['ta'].isna(), 'ta'] = tamean

fall_test.loc[fall_test['ta'].isna(), 'ta'] = tamean
```

```python
#이슬점 온도 
tdmean=fall['td'].mean()
fall.loc[fall['td'].isna(), 'td'] = tdmean

fall_test.loc[fall_test['td'].isna(), 'td'] = tdmean
```

```python
#상대습도
hmmean=fall['hm'].mean()
fall.loc[fall['hm'].isna(), 'hm'] = hmmean

fall_test.loc[fall_test['hm'].isna(), 'hm'] = hmmean
```

```python
#풍속
wsmean=fall['ws'].mean()
fall.loc[fall['ws'].isna(), 'ws'] = wsmean

fall_test.loc[fall_test['ws'].isna(), 'ws'] = wsmean
```
